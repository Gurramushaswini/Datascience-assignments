#99% Confidence Interval Estimation for Print-Head Durability

#Problem Statement

A manufacturer wants to estimate the average durability of print-heads.

Since testing destroys the print-head, only 15 samples were tested.

Using the given sample, we need to construct a 99% confidence interval for the population mean.

Part (a): Use Sample Standard Deviation (t-distribution)

Part (b): Use Population Standard Deviation (Z-distribution)

```python
import numpy as np
data = [1.13,1.55,1.43,0.92,1.25,
        1.36,1.32,0.85,1.07,1.48,
        1.20,1.33,1.18,1.22,1.29]
```

#Calculate Sample Statistics

First, calculate

- Sample Size
- Sample Mean
- Sample Standard Deviation

n = len(data)

mean = np.mean(data)

sample_sd = np.std(data,ddof=1)

print("Sample Size =",n)
print("Sample Mean =",round(mean,4))
print("Sample Standard Deviation =",round(sample_sd,4))

#Part (a)

#99% Confidence Interval using Sample Standard Deviation

Population Standard Deviation is unknown.

Sample Size is less than 30.

Therefore, we use the t-distribution.

Formula

Confidence Interval

Mean ± (t × Standard Error)

Standard Error = Sample SD / √n

```python
t_value = 2.977

standard_error = sample_sd/np.sqrt(n)

margin_error = t_value*standard_error

lower = mean-margin_error

upper = mean+margin_error

print("99% Confidence Interval using t-distribution")

print(round(lower,4),round(upper,4))
```

#Interpretation

We are 99% confident that the true mean durability of the print-heads lies between the calculated lower and upper confidence limits.

#Part (b)

##99% Confidence Interval using Known Population Standard Deviation

Population Standard Deviation is known.

σ = 0.2

Therefore, we use the Z-distribution.

Formula

Confidence Interval

Mean ± (Z × Standard Error)

Standard Error = Population SD / √n

```python
population_sd = 0.2

z_value = 2.576

standard_error = population_sd/np.sqrt(n)

margin_error = z_value*standard_error

lower = mean-margin_error

upper = mean+margin_error

print("99% Confidence Interval using Z-distribution")

print(round(lower,4),round(upper,4))
```

#Interpretation

We are 99% confident that the true population mean durability lies between the calculated confidence limits when the population standard deviation is known.

```python
# Conclusion

In Part (a), the t-distribution was used because the population standard deviation was unknown.

In Part (b), the Z-distribution was used because the population standard deviation was known.

The confidence interval estimates the range in which the true population mean is expected to lie with 99% confidence.
```
