#Hypothesis Testing on Weekly Operating Cost

#Problem Statement

Bombay Hospitality Ltd. uses the following weekly operating cost model:

W = 1000 + 5X

A sample of 25 restaurants reported an average weekly operating cost of Rs.3050.

The objective is to test whether the weekly operating cost is higher than the theoretical cost.

```python
import pandas as pd
sample_mean = 3050

sample_size = 25

x = 600

sigma = 5 * 25

alpha = 0.05
```

#Step 1

Calculate the theoretical mean operating cost.

Formula

W = 1000 + 5X

```python
population_mean = 1000 + 5*x

print("Theoretical Mean =", population_mean)
```


State the Hypotheses

H₀ : μ = 4000

H₁ : μ > 4000

The alternative hypothesis states that the weekly operating cost is higher than the theoretical model.



Calculate the Test Statistic

Formula

Z = (x̄ − μ) / (σ / √n)

```python
import numpy as np
standard_error = sigma / np.sqrt(sample_size)

z = (sample_mean - population_mean) / standard_error

print("Standard Error =", standard_error)

print("Z Statistic =", z)
```

```python
critical_value = 1.645

print("Critical Value =", critical_value)

if z > critical_value:
    print("Reject the Null Hypothesis")
else:
    print("Fail to Reject the Null Hypothesis")
```

#Interpretation

The calculated Z-value is -38.

Since the Z-value is less than the critical value (1.645), we fail to reject the null hypothesis.

#Conclusion
a
There is not enough statistical evidence to support the restaurant owners' claim that the weekly operating cost is higher than the theoretical model.

Therefore, the null hypothesis is not rejected.
