```python
# 1. Create a list of 5 integers. Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse.

numbers = [10, 20, 30, 40, 50]
print("Original List:", numbers)

# append()
numbers.append(60)
print("After append:", numbers)

# extend()
numbers.extend([70, 80])
print("After extend:", numbers)

# insert()
numbers.insert(2, 25)
print("After insert:", numbers)

# remove()
numbers.remove(40)
print("After remove:", numbers)

# pop()
removed_element = numbers.pop()
print("After pop:", numbers)
print("Popped element:", removed_element)

# index()
print("Index of 30:", numbers.index(30))

# count()
numbers.append(20)
print("Count of 20:", numbers.count(20))

# sort()
numbers.sort()
print("After sort:", numbers)

# reverse()
numbers.reverse()
print("After reverse:", numbers)

# clear()
numbers.clear()
print("After clear:", numbers)
```

```python
#11Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then: - Use .info(), .describe() - Select 'Name' and 'City' columns - Drop 'City' column - Fill any missing values in 'Salary' column with the mean - Remove any duplicate row




import pandas as pd

data = {
    "Name": ["Usha", "Ravi", "Sita", "Anil", "Priya", "Kiran", "Deepa", "Rahul", "Sneha", "Ravi"],
    "Age": [22, 25, 23, 24, 26, 21, 27, 22, 23, 25],
    "City": ["Hyderabad", "Chennai", "Bangalore", "Delhi", "Mumbai", "Hyderabad", "Pune", "Delhi", "Chennai", "Chennai"],
    "Salary": [30000, 45000, None, 40000, 50000, 35000, 55000, None, 42000, 45000]
}

df = pd.DataFrame(data)

# 1. Display information
print(df.info())

# 2. Display statistics
print(df.describe())

# 3. Select Name and City columns
print(df[["Name", "City"]])

# 4. Drop City column
df = df.drop("City", axis=1)

# 5. Fill missing Salary values with mean
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())

# 6. Remove duplicate rows
df = df.drop_duplicates()

print(df)
```
