```python

import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("Sales_data_with_discounts.csv")
df
```

```python
# Mode
print(df.mode())
```

```python
import seaborn as sns
for col in df.select_dtypes(include='number').columns:
    plt.figure(figsize=(5,4))
    sns.histplot(df[col], kde=True)
    plt.title(col)
    plt.show()
```

```python
for col in df.select_dtypes(include='number').columns:
    plt.figure(figsize=(5,4))
    plt.boxplot(df[col])
    plt.title(col)
    plt.show()
```

```python
for col in df.select_dtypes(include='object').columns:
    plt.figure(figsize=(5,4))
    df[col].value_counts().plot(kind='bar')
    plt.title(col)
    plt.show()
```

Conclusion:

The Sales and Discounts dataset was analyzed using descriptive analytics and data visualization techniques.
The mean, median, mode, and standard deviation were used to understand the numerical data.
Histograms helped in analyzing the distribution of the data, while boxplots were used to identify outliers and the interquartile range. 
Bar charts were used to visualize the frequency of categorical variables. 
Overall, the analysis provided useful insights into the sales and discount patterns in the dataset.
